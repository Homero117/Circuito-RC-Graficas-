```python
import numpy as np
import matplotlib.pyplot as plt

V0 = 120  # Tensión de la fuente (en voltios)
R = 120  # Resistencia (en ohmios)
C = 0.0002  # Capacidad (en faradios)
tau = R * C

# Arreglo de tiempo
t = np.linspace(0, 2 * tau, 1000)

# Carga del capacitor
Vc_carga = V0 * (1 - np.exp(-t / tau))

# Descarga del capacitor
Vc_descarga = V0 * np.exp(-t / tau)

# Graficar carga y descarga
plt.figure()
plt.plot(t, Vc_carga, label="Carga del Capacitor")
plt.plot(t, Vc_descarga, label="Descarga del Capacitor")
plt.xlabel('Tiempo (segundos)')
plt.ylabel('Tensión en el capacitor (voltios)')
plt.title('Carga y Descarga de un Capacitor en un Circuito RC')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

V0 = 120  # Tensión de la fuente (en voltios)
R = 120  # Resistencia (en ohmios)
C = 0.0002  # Capacidad (en faradios)
tau = R * C
f = 60  # Frecuencia de la señal senoidal (en Hz)

T = 1 / f  # Período de la señal
t = np.linspace(0, 5 * T, 1000)  # Arreglo de tiempo para un ciclo completo

# Tensión en el capacitor en respuesta a una señal senoidal
Vt = V0 * (1 - np.exp(-t / tau))

# Graficar la respuesta a señales periódicas
plt.plot(t, Vt)
plt.xlabel('Tiempo (segundos)')
plt.ylabel('Tensión en el capacitor (voltios)')
plt.title('Respuesta de un Circuito RC a una Señal Senoidal')
plt.grid(True)
plt.show()
```
